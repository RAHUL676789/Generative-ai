# List Comprehension 
## we are going to use looping statement inside the list that we can store  the value inside the list in an iterable and in list comprehension conditonal statement are allowed as well

# syantx for list comprehension
## [expression  for item in iterable if conditon]

### expression ---> what we are going to print


### for item in iterable  ---> looping statemen

### if ---> conditional statement   

```python
lists2 = [x for x in range(1,89)]
print(lists2)
```

# List Comprehendison with conditions
you can add any conditions inside the list comprehension

```python
list_even = [item ** 2 for item in range(1,11) if item % 2 == 0]
print(list_even)
```

## function in list comprehension 
we can also a functioan in list comprehension


```python
list_apple = ["apple" for item in range(1,11) if item % 2 == 0]
print(list_apple)
```

```python
p = ["apple","banana","grapes","mango","orange","kiwi","watermelon"]
```

```python
ls2 = [len(item) for item in p  if len(item) % 2 == 0]
print(ls2)
```

```python
ls1 = [1,2,3]
ls4 = [4,5,6]
ls5 = [[i,j] for i in ls1 for j in ls4]
print(ls5)
```
